1. Crie a imagem de um parabol√≥ide com um eixo dimensionado (como um parabol√≥ide oval)



```python
!pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define os par√¢metros para o parabol√≥ide
a = 1.0  # Fator de escala para o eixo x
b = 0.5  # Fator de escala para o eixo y

# Cria uma grade de pontos usando numpy.linspace
x = np.linspace(-7, 7, 100)
y = np.linspace(-7, 7, 100)
X, Y = np.meshgrid(x, y)

# Calcula os valores de z para o parabol√≥ide
Z = a * X**2 + b * Y**2

# Cria um gr√°fico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Define os r√≥tulos dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Define o t√≠tulo do gr√°fico
plt.title('Parabol√≥ide com Eixos Dimensionados')

# Mostra o gr√°fico
plt.show()
```

2. Crie a imagem de um pecado girado usando rota√ß√£o de coordenadas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define os par√¢metros para o seno girado
frequencia = 2.0  # Frequ√™ncia do seno
angulo_rotacao = np.pi / 4  # √Çngulo de rota√ß√£o (45 graus)

# Cria uma grade de pontos usando numpy.linspace
x = np.linspace(0, 10, 1000)

# Calcula os valores de y para o seno
y = np.sin(2 * np.pi * frequencia * x)

# Aplica a rota√ß√£o de coordenadas
x_rotacionado = x * np.cos(angulo_rotacao) - y * np.sin(angulo_rotacao)
y_rotacionado = x * np.sin(angulo_rotacao) + y * np.cos(angulo_rotacao)

# Plota o seno girado
plt.figure(figsize=(8, 4))
plt.plot(x_rotacionado, y_rotacionado, label='Seno Girado')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Seno Girado com Rota√ß√£o de Coordenadas')
plt.legend()
plt.grid(True)
plt.show()
```

3. Crie a imagem de um gaussiano

```python
import numpy as np
import matplotlib.pyplot as plt

# Define os par√¢metros para a Distribui√ß√£o Gaussiana
media = 0.0
desvio_padrao = 1.0

# Cria uma grade de pontos usando numpy.linspace
x = np.linspace(-5, 5, 1000)

# Calcula os valores da fun√ß√£o de densidade de probabilidade (PDF) da Gaussiana
pdf = (1 / (desvio_padrao * np.sqrt(2 * np.pi))) * np.exp(-(x - media)**2 / (2 * desvio_padrao**2))

# Plota a Distribui√ß√£o Gaussiana
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label='Distribui√ß√£o Gaussiana')
plt.xlabel('X')
plt.ylabel('PDF')
plt.title('Distribui√ß√£o Gaussiana com numpy.linspace()')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define os par√¢metros para a Distribui√ß√£o Gaussiana
media = 0.0
desvio_padrao = 1.0

# Cria uma grade de pontos usando numpy.meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcula os valores da fun√ß√£o de densidade de probabilidade (PDF) da Gaussiana
pdf = (1 / (desvio_padrao * np.sqrt(2 * np.pi))) * np.exp(-(X - media)**2 / (2 * desvio_padrao**2))

# Plota a Distribui√ß√£o Gaussiana
plt.figure(figsize=(8, 4))
plt.contourf(X, Y, pdf, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distribui√ß√£o Gaussiana com numpy.meshgrid()')
plt.colorbar(label='PDF')
plt.grid(True)
plt.show()
```

4. Crie uma fun√ß√£o que gere a imagem de uma Gaussiana opcionalmente rotacionada por um √¢ngulo \theta e com mx, my, sx, sy como argumentos de entrada.

```python
import numpy as np
import matplotlib.pyplot as plt

def rot_gaussiana(mx, my, sx, sy, theta=0):
    """
    Gera uma imagem de uma distribui√ß√£o gaussiana girada.

    Argumentos:
        mx (float): m√©dia ao longo do eixo x.
        my (float): M√©dia ao longo do eixo y.
        sx (float): Desvio padr√£o ao longo do eixo x.
        sy (float): Desvio padr√£o ao longo do eixo y.
        theta (float, opcional): √Çngulo de rota√ß√£o em radianos (o padr√£o √© 0).

    Retorna:
        Nenhum (exibe o gr√°fico).
    """
    # Cria uma grade de pontos usando numpy.meshgrid
    x = np.linspace(-7, 7, 100)
    y = np.linspace(-7, 7, 100)
    X, Y = np.meshgrid(x, y)

    # Calcula os valores da distribui√ß√£o gaussiana
    Z = (1 / (2 * np.pi * sx * sy)) * np.exp(
        -((X - mx) ** 2 / (2 * sx ** 2) + (Y - my) ** 2 / (2 * sy ** 2))
    )

    # Aplica a rota√ß√£o √†s coordenadas
    X_rotacao = X * np.cos(theta) - Y * np.sin(theta)
    Y_rotacao = X * np.sin(theta) + Y * np.cos(theta)

    # Plot da rota√ß√£o gaussiana
    plt.figure(figsize=(8, 4))
    plt.contourf(X_rotacao, Y_rotacao, Z, cmap='viridis')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Roota√ß√£o Gaussiana (Œ∏ = {np.degrees(theta):.2f}¬∞)')
    plt.colorbar(label='PDF')
    plt.grid(True)
    plt.show()

# Exemplo
rot_gaussiana(mx=0, my=0, sx=1, sy=0.5, theta=np.pi / 4)
```
